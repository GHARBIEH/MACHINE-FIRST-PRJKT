```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("data/orders.csv")
order_products_prior = pd.read_csv("data/order_products__prior.csv")
order_products_train = pd.read_csv("data/order_products__train.csv")
products = pd.read_csv("data/products.csv")
aisles = pd.read_csv("data/aisles.csv")
departments = pd.read_csv("data/departments.csv")
```

```python
orders.isnull().sum()
products.isnull().sum()
```

```python
user_orders = orders.groupby("user_id")['order_id'].count().reset_index()
user_orders.columns = ['user_id','total_orders']

product_orders = order_products_prior.groupby("product_id")['order_id'].count().reset_index()
product_orders.columns = ['product_id','total_orders']
```

```python
products_features = products.merge(product_orders, on="product_id", how="left")
products_features.head()
```

```python
products_features.to_csv("data/products_features.csv", index=False)
user_orders.to_csv("data/user_orders.csv", index=False)
```
